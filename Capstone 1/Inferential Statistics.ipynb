```python
# Age
```

```python
sns.set()
sns.distplot(df.age, bins = 20, color = 'orange')
plt.show()
```

```python
# Calculating mean and standard deviation
age = df.age
age_mean, age_std = age.mean(), age.std()
n = len(age)

# ECDF 
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1,n+1)/n
    return x,y
```

```python
# Seeding the random number generator:
np.random.seed(10)

# Drawing random samples from random distribution
age_random_samples = np.random.normal(age_mean, age_std, size=10000)
```

```python
# ECDF on temperature data and on random samples
age_x, age_y = ecdf(age)
age_random_samples_x, age_random_samples_y = ecdf(age_random_samples)

# Plotting the ECDFs for temperature and random sample 
sns.set()
_ = plt.plot(age_random_samples_x, age_random_samples_y)
_ = plt.plot(age_x, age_y, marker = '.', linestyle = 'none')
_ = plt.xlabel('Temperature')
_ = plt.ylabel('CDF')
_ = plt.legend(('Distribution', 'Empirical Data'))
plt.show()
```
